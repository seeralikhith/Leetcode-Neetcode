1. LeetCode #207: Course Schedule <br>
   Difficulty: Medium 

```python
from collections import defaultdict, deque
from typing import List

'''
Possible cases:
    
    1. If numCourses = 1 and len(prerequisites) = any => True.
    
    2. If numCourses = any and len(prerequisites) = 0 => True.
    
    3. If numCourses = any and len(prerequisites) = any => True (If no cycle and self-loop).
        
        3.1. If Ai != Bi and Ai -> Bi and !Bi -> Ai or Ai <- Bi and Bi <- !Ai => True
    
    4. If numCourses = any and len(prerequisites) = any => False (If cycle and self-loop)
        
        4.1. If Ai == Bi => Self-Loop => False [in prerequisite]
        
        4.2 If Ai != Bi and If Ai -> Bi and Bi -> Ai => Cycle => False [in prerequisite]
'''

class Solution:
    def canFinish(self, numCourses: int, prerequisites: List[List[int]]) -> bool:
        # print(numCourses, prerequisites)
        if numCourses == 1 or len(prerequisites) == 0:      
            # Cases 1 and 2
            return True

        else:       
            # Cases 3 and 4
            '''
                1. prerequisites is in the form of "List of lists". 
                2. In prerequisites, If you need Ai, you must complete Bi first.
                3. Required is if user can complete all numCourses or not.

                This is problem is similar to Topo Sort.
            '''
            
            graphAdjList = defaultdict(list)
            # If we are not sure of keys, and values are fixed list, we use defaultdict for graph adjacent list.

            inDegree = {i : 0 for i in range(0, numCourses)}
            
            for a_i, b_i in prerequisites:
                # b_i --> a_i
                graphAdjList[b_i].append(a_i)       
                    
                # Course a_i will have in-degree
                inDegree[a_i] += 1
            
            # print(graphAdjList)

            '''
            Topological Sort using BFS
            '''
            # Declare a queue.
            graphQ = deque()

            # In this queue, we insert vertices with zero in-degree
            for course, id_count in inDegree.items():
                if id_count == 0:
                    graphQ.append(course)
                else:
                    continue

            # Let us assume a count variable for the vertices which are visited.
            # and a list variable to store the order of the vertices.
            course_count = 0
            # courses_order = []

            while graphQ:
                vertex = graphQ.popleft()
                
                # The indegree = 0 vertex from queue is visited. 
                # So, count is incremented and vertex is stored in sort list.
                course_count += 1
                # courses_order.append(vertex)    
                
                # Visiting all the neighbor nodes of current node using Adj. list
                for neighbor in graphAdjList[vertex]:

                    # Since the neighbor is visited, the inDegree is reduced by 1.
                    inDegree[neighbor] -= 1

                    # Inserting back the neighbor vertices with zero In-degree.
                    if inDegree[neighbor] == 0:
                        graphQ.append(neighbor)
            
            # If user is able to take all courses, return True
            # Condition: There should be no cycle or self-loop
            if course_count == numCourses:
                # print('Topo Sort Order:',courses_order)
                return True
            else:                   # If course_count != numCourses
                # There exists cycle or self-loop 
                return False


if __name__ == "__main__":

    solution = Solution()

    # Example-1
    numCourses = 2
    prerequisites = [[1,0]]
    # Output: True

    result = solution.canFinish(numCourses, prerequisites)
    print(result)

    # Example-2
    numCourses = 2
    prerequisites = [[1,0],[0,1]]
    # Output: False

    solution = Solution()
    
    result = solution.canFinish(numCourses, prerequisites)
    print(result)  
```

```python
from collections import defaultdict, deque
from typing import List

'''
Possible cases:
    
    1. If numCourses = 1 and len(prerequisites) = 0 => [0].
    
    2. If numCourses = 1 and len(prerequisites) = any => [] (Invalid Case).

    3. If numCourses = any(n) and len(prerequisites) = 0 => [0,1,2,... (n-1)] . 
    
    4. If numCourses = any and len(prerequisites) = any => return courses_order (If no cycle and self-loop).
        
        4.1. If Ai != Bi and Ai -> Bi and !Bi -> Ai or Ai <- Bi and Bi <- !Ai => return courses_order
    
    5. If numCourses = any and len(prerequisites) = any => return [] (If cycle and self-loop)
        
        5.1. If Ai == Bi => Self-Loop => return [] [in prerequisite]
        
        5.2 If Ai != Bi and If Ai -> Bi and Bi -> Ai => Cycle => return [] [in prerequisite]
'''

class Solution:
    def findOrder(self, numCourses: int, prerequisites: List[List[int]]) -> List[int]:
        # print(numCourses, prerequisites)
        if numCourses == 1 and len(prerequisites) == 0:      
            # Case 1
            return [0]

        elif numCourses == 1 and len(prerequisites) > 0:      
            # Case 2
            return []
        
        elif numCourses > 1 and len(prerequisites) == 0:
            # Case 3
            return [i for i in range(0, numCourses)]
        
        else:       
            # Cases 4 and 5
            '''
                1. prerequisites is in the form of "List of lists". 
                2. In prerequisites, If you need Ai, you must complete Bi first.
                3. Required is if user can complete all numCourses or not.

                This is problem is similar to Topo Sort.
            '''
            
            graphAdjList = defaultdict(list)
            # If we are not sure of keys, and values are fixed list, we use defaultdict for graph adjacent list.

            inDegree = {i : 0 for i in range(0, numCourses)}
            
            for a_i, b_i in prerequisites:
                # b_i --> a_i
                graphAdjList[b_i].append(a_i)       
                    
                # Course a_i will have in-degree
                inDegree[a_i] += 1
            
            # print(graphAdjList)

            '''
            Topological Sort using BFS
            '''
            # Declare a queue.
            graphQ = deque()

            # In this queue, we insert vertices with zero in-degree
            for course, id_count in inDegree.items():
                if id_count == 0:
                    graphQ.append(course)
                else:
                    continue

            # Let us assume a count variable for the vertices which are visited.
            # and a list variable to store the order of the vertices.
            # course_count = 0
            courses_order = []

            while graphQ:
                vertex = graphQ.popleft()
                
                # The indegree = 0 vertex from queue is visited. 
                # So, count is incremented and vertex is stored in sort list.
                # course_count += 1
                courses_order.append(vertex)    
                
                # Visiting all the neighbor nodes of current node using Adj. list
                for neighbor in graphAdjList[vertex]:

                    # Since the neighbor is visited, the inDegree is reduced by 1.
                    inDegree[neighbor] -= 1

                    # Inserting back the neighbor vertices with zero In-degree.
                    if inDegree[neighbor] == 0:
                        graphQ.append(neighbor)
            
            # If user is able to take all courses, return True
            # Condition: There should be no cycle or self-loop
            if len(courses_order) == numCourses:
                # print('Topo Sort Order:',courses_order)
                return courses_order
            else:                   # If course_count != numCourses
                # There exists cycle or self-loop 
                return []


if __name__ == "__main__":

    solution = Solution()

    # Example-1
    numCourses = 2
    prerequisites = [[1,0]]
    # Output: True

    result = solution.findOrder(numCourses, prerequisites)
    print(result)

    # Example-2
    numCourses = 2
    prerequisites = [[1,0],[0,1]]
    # Output: False
    
    result = solution.findOrder(numCourses, prerequisites)
    print(result)

    numCourses = 4
    prerequisites = [[1,0],[2,0],[3,1],[3,2]]

    result = solution.findOrder(numCourses, prerequisites)
    print(result)

    numCourses = 1
    prerequisites = [[1,0]]

    result = solution.findOrder(numCourses, prerequisites)
    print(result)  

    numCourses = 1
    prerequisites = []

    result = solution.findOrder(numCourses, prerequisites)
    print(result)  
```

3. LeetCode #269: Alien Dictionary<br>

##### Method-1: Using Iterative BFS Topo Sort.

```python
from collections import defaultdict, deque
from typing import List

'''
Possible cases:
    
    1. If len(words) == 1 => . 
        1.1 If len(words) == 1 and len(words[0]) == 1
        1.2 If len(words) == 1 and len(words[0]) == any 
    
    2. If len(words[i]) == 1 => .
        2.1 If len(words) == any and len(words[i]) == 1

    3. If len(words) >= 2 =>
            first word = words[0]
            second word = words[1:]

        3.1  If len(first word) > len(second word) => Invalid case
        3.2  If len(first word) =< len(second word) => Valid case
    
    4.  Repeated Characters => Select the repeated letter once only.    ==> Done using Sets
    
    5. Check if there are different swapped values as they form cycles and self-loops.
'''

class Solution:
    def alienOrder(self, words: List[str]) -> str:
        # print(words[1:])
        if len(words) == 0:
            return ""
        
        elif len(words) == 1: 
            # Case 1: Only one word     
            wordSet = sorted(set(words[0]))
            return "".join(wordSet)
        
        else:
            # Case 2: Multiple words

            '''
                1. words is in the form of "List". 
                2. In words, some letters are connected as u comes before v.
                3. Required is that the alien order exists or not.

                This is problem is similar to Topo Sort.
            '''
            # Step 1: Initialize graph and in-degree

            graphAdjList = defaultdict(set)
            # If we are not sure of keys, and values are fixed set, we use defaultdict for graph adjacent list. 
            # Sets are used to remove any repeated letters ==> Case 4

            inDegree = {char: 0 for word in words for char in word}

            # Step 2: Build the graph and in-degree       
            for word1, word2 in zip(words, words[1:]):
                # print('A', word1, word2)
                # Case 3.2

                min_length = min(len(word1), len(word2))

                # Check for invalid case where prefix is longer than the word following it
                if len(word1) > len(word2) and word1[:min_length] == word2[:min_length]:
                        return ""       # Case 3.1 Invalid case: Length mismatch
                else:
                    for char1, char2 in zip(word1, word2):
                        # print('B', char1, char2)
                        if char1 == char2:
                            continue
                        else:           # char1 != char2
                            if char2 not in graphAdjList[char1]:
                                graphAdjList[char1].add(char2)
                                inDegree[char2] += 1
                            break
            
            # print(graphAdjList)
            # print(inDegree)
            '''
            Topological Sort using BFS : Kahn's Algorithm
            '''

            
            # Step 3: Perform topological sorting using DFS

            # Declare a queue.
            graphQ = deque()

            # In this queue, we insert vertices with zero in-degree 
            for charIn, id_count in inDegree.items():
                if id_count == 0:
                    graphQ.append(charIn)
                else:
                    continue
    
            # print(graphQ)

            # Let us assume a count variable for the vertices which are visited.
            # and a list variable to store the order of the vertices.
            # alien_count = 0
            alien_order = []

            while graphQ:
                vertex = graphQ.popleft()
                
                # The indegree = 0 vertex from queue is visited. 
                # So, count is incremented and vertex is stored in sort list.
                # alien_count += 1
                alien_order.append(vertex)    
                
                # Visiting all the neighbor nodes of current node using Adj. list
                for neighbor in graphAdjList[vertex]:

                    # Since the neighbor is visited, the inDegree is reduced by 1.
                    inDegree[neighbor] -= 1

                    # Inserting back the neighbor vertices with zero In-degree.
                    if inDegree[neighbor] == 0:
                        graphQ.append(neighbor)

            # print(alien_order)
           
            # Condition: There should be no cycle or self-loop
            # Check if valid alien order exists
            if len(alien_order) == len(inDegree):
                return "".join(map(str, alien_order))
            else:
                # There exists cycle or self-loop 
                # Case 5 and Alien order doesn't exist
                return ""

if __name__ == "__main__":

    solution = Solution()

    # Example-1
    words = ["wrt","wrf","er","ett","rftt"]
    # Output: True

    result = solution.alienOrder(words)
    print(result)

    # Example-2
    words = ["z","x"]
    # Output: False
    
    result = solution.alienOrder(words)
    print(result)

    words = ["z","x","z"]

    result = solution.alienOrder(words)
    print(result)

    words = ["wrt","wrf","er","ett","rftt","te"]

    result = solution.alienOrder(words)
    print(result)
  
```

##### Method-2: Using Recursive DFS Topo Sort.

```python

from typing import List, Dict, Set
from collections import defaultdict

'''
Possible cases:
    
    1. If len(words) == 1 => . 
        1.1 If len(words) == 1 and len(words[0]) == 1
        1.2 If len(words) == 1 and len(words[0]) == any 
    
    2. If len(words[i]) == 1 => .
        2.1 If len(words) == any and len(words[i]) == 1

    3. If len(words) >= 2 =>
            first word = words[0]
            second word = words[1:]

        3.1  If len(first word) > len(second word) => Invalid case
        3.2  If len(first word) =< len(second word) => Valid case
    
    4.  Repeated Characters => Select the repeated letter once only.    ==> Done using Sets
    
    5. Check if there are different swapped values as they form cycles and self-loops.
'''

class Solution:
    def alienOrder(self, words: List[str]) -> str:
        # print(words[1:])
        if len(words) == 0:
            return ""
        
        elif len(words) == 1: 
            # Case 1: Only one word     
            wordSet = sorted(set(words[0]))
            return "".join(wordSet)
        
        else:
            # Case 2: Multiple words

            '''
                1. words is in the form of "List". 
                2. In words, some letters are connected as u comes before v.
                3. Required is that the alien order exists or not.

                This is problem is similar to Topo Sort.
            '''
            # Step 1: Initialize graph and in-degree

            graphAdjList = defaultdict(set)
            # If we are not sure of keys, and values are fixed set, we use defaultdict for graph adjacent list. 
            # Sets are used to remove any repeated letters ==> Case 4

            inDegree = {char: 0 for word in words for char in word}

            # Step 2: Build the graph and in-degree       
            for word1, word2 in zip(words, words[1:]):
                # print('A', word1, word2)
                # Case 3.2

                min_length = min(len(word1), len(word2))

                # Check for invalid case where prefix is longer than the word following it
                if len(word1) > len(word2) and word1[:min_length] == word2[:min_length]:
                        return ""       # Case 3.1 Invalid case: Length mismatch
                else:
                    for char1, char2 in zip(word1, word2):
                        # print('B', char1, char2)
                        if char1 == char2:
                            continue
                        else:           # char1 != char2
                            if char2 not in graphAdjList[char1]:
                                graphAdjList[char1].add(char2)
                                inDegree[char2] += 1
                            break
            
            # print(graphAdjList)
            # print(inDegree)

            '''
            Topological Sort using DFS : Recursive Approach
            Topological Sort using BFS (Kahn's Algorithm): Iterative Approach
            
            '''
            # Step 3: Perform topological sorting using DFS

            # Recursive approach uses a stack version of data structure which is implicit.

            visited = {char: 0 for char in inDegree}        
            # A dictionary to store the node(char) status; whether it is unvisited--> 0, visiting--> 1 or visited--> 2
            # print(visited)

            # Let us assume a count variable for the vertices which are visited.
            # and a list variable to store the order of the vertices.
            # alien_count = 0
            alien_order = []

            def topoSortDFS(node: str, graphAdjList: Dict[str, Set[str]], stack: List[str], visited: Dict[str, int]) -> bool:

                if visited[node] == 1:          # Checks if the current node is being visited in the current traversal.
                    return False                # If yes, a Cycle detected
                
                if visited[node] == 2:      # Checks if the current node has already been explored and added to the stack.
                    return True             # If yes, the node is explored and skip further traversal
                
                visited[node] = 1           # Prevents processing the same node again in the current traversal to avoid infinite recursion.
                
                # Visiting all the neighbor nodes of current node using Adj. list
                if node in graphAdjList:
                    for neighbor in graphAdjList[node]:
                        if topoSortDFS(neighbor, graphAdjList, stack, visited) == False:
                            # If any neighbor returns False, topoSortDFS(node) also returns False, indicating a cycle.
                            return False

                visited[node] = 2       # Indicates that the node along with all its neighbors are visited. 
                stack.append(node)      # Added to the stack for representing in the Topo Sort order.

            
                return True         # If the whole process is complete, it returns True.
            
            for charIn, id_count in inDegree.items():
                if id_count == 0 and not topoSortDFS(charIn, graphAdjList, alien_order, visited):
                    return ""

            # print(alien_order)
            if len(alien_order) != len(inDegree):
                # There exists cycle or self-loop 
                # Case 5: # Incomplete topological sorting due to cycles ==> Alien order doesn't exist
                return ""  
            else:
                # Condition: There should be no cycle or self-loop
                # Check if valid alien order exists
                return "".join(alien_order[::-1])

            
if __name__ == "__main__":
    solution = Solution()

    # Test cases
    words1 = ["wrt","wrf","er","ett","rftt"]
    print(solution.alienOrder(words1))  # Output: "wertf"

    words2 = ["z","x"]
    print(solution.alienOrder(words2))  # Output: "zx"

    words3 = ["z","x","z"]
    print(solution.alienOrder(words3))  # Output: ""

    words4 = ["wrt","wrf","er","ett","rftt","te"]
    print(solution.alienOrder(words4))  # Output: "wertf"
```

4. LeetCode #277. Find the Celebrity<br>
   Difficulty: Medium 

Based on `knows(A,B) = "A knows B"`:

#### Problem Analysis
1. **If A knows B:**
   - "B may or may not be a celebrity"
   - "A is not the celebrity" (because a celebrity doesn't know anyone)

   This implies:
   - `knows(A,B) = True` ⇒ A is not the celebrity (Sure statement)

2. **If A doesn't know B:**
   - "A may or may not be a celebrity"
   - "B is not the celebrity" (because everyone knows the celebrity)

   This implies:
   - `knows(A,B) = False` ⇒ B is not the celebrity (Sure statement)

#### Approach to Find the Celebrity
Using the above statements, we can repeatedly use `knows(A,B)` to eliminate non-celebrities from the group, leaving only one potential celebrity.

#### Possible Approach for `knows(A,B)`
Let there be nodes from 0 to (N-1), a total of N nodes.

1. **Base Case:**
   - If A = 0 and B = 1:
     - `knows(0,1) = True`: Node 0 knows Node 1 ⇒ Node 0 is not a celebrity ⇒ Node 1 may be the potential candidate.
     - `knows(0,1) = False`: Node 0 doesn't know Node 1 ⇒ Node 1 is not a celebrity ⇒ Node 0 may be the potential candidate.
   - If A = 1 and B = 0:
     - `knows(1,0) = True`: Node 1 knows Node 0 ⇒ Node 1 is not a celebrity ⇒ Node 0 may be the potential candidate.
     - `knows(1,0) = False`: Node 1 doesn't know Node 0 ⇒ Node 0 is not a celebrity ⇒ Node 1 may be the potential candidate.

2. **Iterative/Recursive Case:**
   Using the base case, repeat this process of comparison until the (n-1) nodes:
   - For loop where 'i' traverses from the 2nd to the (n-1)th node:
     1. If `knows(candidate, i) == True`:
        - Node candidate knows Node i ⇒ Node candidate is not a celebrity ⇒ Node i may be the potential candidate.
        - (Set `candidate = i`)
     2. If `knows(candidate, i) == False`:
        - Node candidate doesn't know Node i ⇒ Node i is not a celebrity ⇒ Node candidate may continue to be the potential candidate.
        - (Continue with the current candidate)

3. **Terminating Case:**
   From both cases, check:
   1. If `knows(candidate, k) == True` or `knows(k, candidate) == False`:
      - Node candidate knows Node k or Node k doesn't know Node candidate ⇒ Node 'k' is not the celebrity.
   2. If `knows(candidate, k) == False` and `knows(k, candidate) == True`:
      - Node candidate doesn't know Node k and Node k knows Node candidate ⇒ Node 'candidate' is the celebrity.

##### Method-1: Using in-Degrees and out-Degrees of a Graph : Time Limit Exceeded

```python
'''
Method - 1:
This problem is similar to determining the in-degree and out-degree of a directed graph. 
Here, the celebrity is a node with maximum in-degree and zero out-degree.
'''             

# The knows API is already defined for you.
# return a bool, whether a knows b

# def knows(a: int, b: int) -> bool:
#     pass

from typing import List, Dict, Set

class Solution:
    def findCelebrity(self, n: int) -> int:
        # print(n, knows(0,0))

        inDegrees = {node: 0 for node in range(0,n)}
        outDegrees = {node: 0 for node in range(0,n)}
        
        for i in range(0,n):
            for j in range((i+1),n):            
                if knows(i,j) == 1:
                    # Node i knows Node j
                    inDegrees[j] += 1
                    outDegrees[i] += 1
            
                if knows(j,i) == 1:
                    # Node j knows Node i
                    inDegrees[i] += 1
                    outDegrees[j] += 1

        # print(inDegrees, outDegrees)

        celebrity = -1      # Let us assume there is no celebrity node index present. 
        for k in range(0, n):
            if inDegrees[k] == n-1 and outDegrees[k] == 0:
                celebrity = k

        return celebrity       

if __name__ == "__main__":
    graph = [[1, 1, 0], [0, 1, 0], [1, 1, 1]]
    def knows(a, b):
        return graph[a][b]

    solution = Solution()
    print(solution.findCelebrity(len(graph)))  # Output: 1
        
```

##### Method-2: Using Candidate Verification and Candidate Validation: (Based on Two-Pointers and Logic of the problem)

```python
'''
Method - 2:
This problem can be solved using two-pointer concept. 
Here, celebrity is the index of that pointer variable which satisfies all the conditions of the problem.
'''             

# The knows API is already defined for you.
# return a bool, whether a knows b

# def knows(a: int, b: int) -> bool:
#     pass

from typing import List, Dict, Set

class Solution:
    def findCelebrity(self, n: int) -> int:
        # print(n, knows(0,0))

        # Step 1: Candidate Identification : Two Pointers
        a = 0
        b = n-1
        while a < n and b < n:
            if knows(a,b) == True:  
                # Node a knows Node b => Node a is not a celebrity. 
                # Move to next value.
                a += 1
            else:           # knows(a,b) == False  
                # Node a doesn't know Node b => Node b is not a celebrity.
                # Move to previous value. 
                b -= 1
            
            if a >= b:
                # To avoid repetition of visiting of nodes.
                break

        celebrity = a
        
        # Step 2: Candidate Verification : Problem Logic
        for j in range(0,n):
            if j != celebrity and knows(celebrity, j) == True or knows(j, celebrity) == False:
                return -1
            
        return celebrity
        
if __name__ == "__main__":
    
    def knows(a, b):
        return graph[a][b]

    solution = Solution()

    graph = [[1, 1, 0], [0, 1, 0], [1, 1, 1]]
    print(solution.findCelebrity(len(graph)))  # Output: 1

    graph = [[1,0,1],[1,1,0],[0,1,1]]
    print(solution.findCelebrity(len(graph)))   # Output: -1
```

5. LeetCode #329. Longest Increasing Path in a Matrix<br>
   Difficulty: Hard

##### Method-1: Using DAGs and TopoSort:

The problem is in the form of a matrix, if we consider it as an adjacent matrix. Then the rows and columns can represent the nodes and edges of the graph.
So, the required is Longest increasing path, so we can use Topological sort for this. As, it is Directed Acyclic graph and we need to travel each and every node as in the go, so this concept could be used.

```python
from typing import List
from collections import deque

class Solution:
    def longestIncreasingPath(self, matrix: List[List[int]]) -> int:
        if not matrix or not matrix[0]:
            return 0
        
        R = len(matrix)
        C = len(matrix[0])

        # Possible movements allowed: [Left, Right, Down, Up]
        directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]     

        # Step 1: Initialize graph and in-degree
        inDegree = {(row, column): 0 for row in range(R) for column in range(C)}
        graphAdjList = {(row, column): [] for row in range(R) for column in range(C)}

        # Step 2: Build the graph and in-degree
        for row in range(R):
            for column in range(C):
                for (x, y) in directions:
                    rx = row + x
                    cy = column + y

                    if rx < 0 or rx >= R or cy < 0 or cy >= C:
                        continue
                    
                    if 0 <= rx < R and 0 <= cy < C:
                        # (rx, cy) <--- (row, column)   Creating a connection
                        if matrix[row][column] < matrix[rx][cy]:
                            graphAdjList[(row, column)].append((rx, cy))
                            inDegree[(rx, cy)] += 1
        
        '''
        Topological Sort using BFS : Kahn's Algorithm
        '''
            
        # Step 3: Perform topological sorting using BFS

        # Declare a queue.
        graphQ = deque()

        # In this queue, we insert vertices with zero in-degree 
        for coords, id_count in inDegree.items():
            if id_count == 0:
                graphQ.append(coords)

        # Let us assume a count variable for the vertices which are visited.
        # and a list variable to store the order of the vertices.
        
        # Here, the main agenda is to find out what is the longest increasing path, 
        # so the conditions of cycle is out of equation.
        maxPathCount = 0

        while graphQ:
            sizeQ = len(graphQ)
            i = 0
            temp = deque()

            while i < sizeQ:
                row, col = graphQ.popleft()

                # Visiting all the neighbor nodes of current node using Adj. list
                for r, c in graphAdjList[(row, col)]:
                    # Since the neighbor is visited, the inDegree is reduced by 1.
                    inDegree[(r, c)] -= 1

                    # Inserting back the neighbor vertices with zero In-degree.
                    if inDegree[(r, c)] == 0:
                        temp.append((r, c))
                i += 1

            graphQ.extend(temp)
            maxPathCount += 1

        return maxPathCount      
    
if __name__ == "__main__":
    solution = Solution()

    matrix = [[9, 9, 4], [6, 6, 8], [2, 1, 1]]
    print(solution.longestIncreasingPath(matrix))  # Output: 4

    matrix = [[3, 4, 5], [3, 2, 6], [2, 2, 1]]
    print(solution.longestIncreasingPath(matrix))  # Output: 4

    matrix = [[1]]
    print(solution.longestIncreasingPath(matrix))  # Output: 1
```

##### Method-2: Using DFS with Memoization (Dynamic Programming):

The problem is in the form of a matrix, if we consider it as an adjacent matrix. Then the rows and columns can represent the nodes and edges of the graph.
So, the required is Longest increasing path, so we can use traversal for this. As we need to travel each and every node as in the go, DFS would be a better preferrence.

Also, for better performance, we would be using memoization concept from dynamic programming to keep track of count of connecting nodes in the path at every node and return the maximum number of nodes connecting the path.

To get the path, from the node with the maximum node connecting if we move to the nodes with decreasing count within the directions, we can retrive it.


```python
from typing import List, Dict, Set

class Solution:
    def longestIncreasingPath(self, matrix: List[List[int]]) -> int:
        if not matrix or not matrix[0]:
            return 0
        
        R = len(matrix)
        C = len(matrix[0])

        dp = {}         # This dictionary is used to store the cells present in the longest path.
                        # If you re-encounter a cell, instead of recalculating, you directly retrieve the length of the longest path starting from that cell.
                        # It is stored in the format as (row# , column#) --> count, Longest Increasing Path.

        def traversalDFS(row: int, column: int) -> int:
            if row < 0 or row >= R or column < 0 or column >= C:
                return 0
            
            curVal = matrix[row][column]
            
            if (row, column) in dp:             # If already visited, refer to dp for value retrival.
                return dp[(row, column)]

            if row > 0 and matrix[row-1][column] > curVal:       
                leftDFS = traversalDFS(row-1, column)
            else:
                leftDFS = 0

            if column > 0 and matrix[row][column-1] > curVal:
                downDFS = traversalDFS(row, column-1)
            else:
                downDFS = 0

            if row < R-1 and matrix[row+1][column] > curVal:
                rightDFS = traversalDFS(row+1, column)
            else:
                rightDFS = 0

            if column < C-1 and matrix[row][column+1] > curVal:
                upDFS = traversalDFS(row, column+1)
            else:
                upDFS = 0

            # Storing the path
            maxPath = 1 + max(leftDFS, rightDFS, upDFS, downDFS)         # 1 + .. ==> 1 includes the assumption of curVal = matrix[row][column]

            dp[(row, column)] = maxPath
            return maxPath
            
        # Traversing every value in the matrix.
        for r in range(0, R):
            for c in range(0, C):
                traversalDFS(r, c)

        # print(dp)                 

        # Longest Increasing Path is nothing but the maximum value from the dp hashtable.
        return max(dp.values())
    

if __name__ == "__main__":

    solution = Solution()

    matrix = [[9,9,4],[6,6,8],[2,1,1]]
    print(solution.longestIncreasingPath(matrix))  # Output: 4

    matrix = [[3,4,5],[3,2,6],[2,2,1]]
    print(solution.longestIncreasingPath(matrix))   # Output: 4

    matrix = [[1]]
    print(solution.longestIncreasingPath(matrix))   # Output: 1
```
