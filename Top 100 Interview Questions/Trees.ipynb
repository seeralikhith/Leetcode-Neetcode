1. LeetCode #94: Binary Tree Inorder Traversal <br>
   Difficulty: Easy

```python
from typing import Optional, List

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

# Recursive method
# class Solution:
#     def inorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
        
#         def inOrderDFS(node, seq):
#             if node is not None:
#                 if node.left is not None:
#                     inOrderDFS(node.left, seq)
                
#                 if node.val is not None:
#                     seq.append(node.val)

#                 if node.right is not None:
#                     inOrderDFS(node.right, seq)
                
#                 return seq
#             else:
#                 return []

#         res = []
#         inOrderDFS(root, res)
#         return res

# Iterative Method
class Solution:
    def inorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
        seq = []

        if root is None:
            return []

        else:
            # print(root)
            stack = [(root, False)]

            while stack:
                    node, isVisited = stack.pop()
                    # print('A', node.val)

                    if node is not None:
                        if isVisited == True:
                            if node.val is not None:
                                seq.append(node.val)
                        else:
                            # inorder: left -> root -> right 
                            # But in stack inorder is stored as: right -> root -> left
                            if node.left is not None:
                                stack.append((node.left, False))

                            stack.append((node, True))                


                            if node.right is not None:
                                stack.append((node.right, False))
                        
                    else:
                        continue
        
        # So in the result, we reverse the list for inorder traversal
        return seq[::-1]

  

if __name__ == "__main__":
    # Tree 1
    # p = [1,2,2,3,4,4,3]
    # p = TreeNode(1)
    # p.left = TreeNode(2)
    # p.right = TreeNode(2)
    # p.left.left = TreeNode(3)
    # p.left.right = TreeNode(4)
    # p.right.left = TreeNode(4)
    # p.right.right = TreeNode(3)

    # Tree 2
    # p = [1,null,2,3]
    # p = TreeNode(1)
    # p.left = TreeNode(None)
    # p.right = TreeNode(2)
    # p.right.left = TreeNode(3)

    # Tree 3
    # p = []
    # p = TreeNode(None)
    
    # Tree 4
    p = [1]
    p = TreeNode(1)
    
    # Create an instance of Solution
    solution = Solution()
    
    # Check if the trees are the same
    result = solution.inorderTraversal(p)
    print(result)
```

3. LeetCode #101: Symmetric Tree<br>

```python
from typing import Optional
  
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def isSymmetric(self, root: Optional[TreeNode]) -> bool:
        # print(root.val)
        flag = False

        if root is None:
            flag = True

        else:
            flag = self.isReflection(root.left, root.right)

        return flag
        
    def isReflection(self, n_left: Optional[TreeNode], n_right: Optional[TreeNode]) -> bool:
        flag = False

        if n_left is None and n_right is None:
            flag = True
                    
        elif (n_left is None and n_right is not None) or (n_left is not None and n_right is None):
            flag = False
        
        else:           # p is not None and q is not None
            if (n_left.val == n_right.val) and self.isReflection(n_left.left, n_right.right) and self.isReflection(n_left.right, n_right.left):
                flag = True
            else:
                flag = False
        
        return flag

# @lc code=end
if __name__ == "__main__":
    # Tree 1
    # p = [1,2,2,3,4,4,3]
    p = TreeNode(1)
    p.left = TreeNode(2)
    p.right = TreeNode(2)
    p.left.left = TreeNode(3)
    p.left.right = TreeNode(4)
    p.right.left = TreeNode(4)
    p.right.right = TreeNode(3)
    
    # [1,2,2,null,3,null,3]
    # p = TreeNode(1)
    # p.left = TreeNode(2)
    # p.right = TreeNode(2)
    # p.left.left = TreeNode(None)
    # p.left.right = TreeNode(3)
    # p.right.left = TreeNode(None)
    # p.right.right = TreeNode(3)
    # # Create an instance of Solution
    solution = Solution()
    
    # Check if the trees are the same
    result = solution.isSymmetric(p)
    print(result)
```

```python
from typing import List, Optional

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
        
class Solution:
    def buildTree(self, inorder: List[int], postorder: List[int]) -> Optional[TreeNode]:
        if not inorder or not postorder:
            return None

        inorderMap = {value: index for index, value in enumerate(inorder)}  # Store inorder indices in a dictionary
        
        def treeBuilder(treeOrder1, start1, end1, treeOrder2, start2, end2):
            # Base case: If there are no elements to consider
            if start1 > end1 or start2 > end2:
                return None
            
            elif start1 == end1 or start2 == end2:
                return TreeNode(treeOrder1[end1])

            else:
                # Last element in postorder is the root
                rootValue = treeOrder1[end1]
                rootSubTree = TreeNode(rootValue)
                
                # Root index in inorder traversal
                rootIndex = inorderMap[rootValue]
                
                # Number of elements in left subtree
                numLeftChild = rootIndex - start2
                
                # Recursively build the left subtree
                rootSubTree.left = treeBuilder(treeOrder1, start1, start1 + numLeftChild - 1, treeOrder2, start2, rootIndex - 1)
                
                # Recursively build the right subtree
                rootSubTree.right = treeBuilder(treeOrder1, start1 + numLeftChild, end1 - 1, treeOrder2, rootIndex + 1, end2)
                
                return rootSubTree
        
        return treeBuilder(postorder, 0, len(postorder) - 1, inorder, 0, len(inorder) - 1)
```

```python
from typing import Optional, List
import collections
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

# Recursive InOrder
# class Solution:
#     def kthSmallest(self, root: Optional[TreeNode], k: int) -> int:
#         def inOrderDFS(node, seq):
#                 if node is not None:
#                     if node.left is not None:
#                         inOrderDFS(node.left, seq)
                    
#                     if node.val is not None:
#                         seq.append(node.val)

#                     if node.right is not None:
#                         inOrderDFS(node.right, seq)
                    
#                     return seq
#                 else:
#                     return []

#         res = inOrderDFS(root, [])
#         # print(res[k-1])
#         return res[k-1]

# Iterative Inorder
class Solution:
    def kthSmallest(self, root: Optional[TreeNode], k: int) -> int:
        if root is None:
            return []
        
        stack = [root]
        curr = root.left
        result = []

        while (curr is not None) or (len(stack) > 0):
            # Reach the leftmost TreeNode
            while curr is not None:
                stack.append(curr)
                curr = curr.left

            # print([i.val for i in stack])

            # Reached leftmost point
            curr = stack.pop()
            result.append(curr.val)

            # Visit rightsub-tree
            curr = curr.right

        # print(result)
        return result[k-1]     
        
        


if __name__ == "__main__":
    # Tree 1
    # p = [5,3,6,2,4,null,null,1]
    p = TreeNode(5)
    p.left = TreeNode(3)
    p.right = TreeNode(6)
    p.left.left = TreeNode(2)
    p.left.right = TreeNode(4)
    p.left.left.left = TreeNode(1)

    # root = [5,3,6,2,4,None,None,1]
    k = 3

    # Create an instance of Solution
    solution = Solution()
    
    # Check if the trees are the same
    result = solution.kthSmallest(p, k)
    print(result)        
```
