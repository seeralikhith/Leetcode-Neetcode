1. LeetCode #704: Binary Search <br>
   Difficulty: Easy 

```python
from typing import List

# Recursive Method
class Solution:
    def search(self, nums: List[int], target: int) -> int:
        high = 0
        low = len(nums) - 1

        def binarySearch(nums, start, end, target):
            if start > end:
                return -1
            
            while start <= end:
                mid = (start + end) // 2
                if nums[mid] == target:
                    return mid
                elif nums[mid] < target:
                    return binarySearch(nums, mid + 1, end, target)
                else:       # nums[mid] > target
                    return binarySearch(nums, start, mid - 1, target)

        return binarySearch(nums, high, low, target)

# # Iterative Method
# class Solution:
#     def search(self, nums: List[int], target: int) -> int:
#         start = 0
#         end = len(nums) - 1

#         while start <= end:
#             mid = (start + end) // 2
#             if nums[mid] == target:
#                 return mid
#             elif nums[mid] < target:
#                 start = mid + 1
#             else:       # nums[mid] > target
#                 end = mid - 1

#         if start > end:
#             return -1
        
if __name__ == "__main__":

    solution = Solution()

    # Example-1
    nums = [-1,0,3,5,9,12]
    target = 2
    # Output: -1

    result = solution.search(nums, target)
    print(result)

    # Example-2
    nums = [-1,0,3,5,9,12]
    target = 9
    # Output: 4
    solution = Solution()
    
    result = solution.search(nums, target)
    print(result)  
```

5. LeetCode #74: Search a 2-D Matrix <br>
   Difficulty: Medium 

```python
from typing import List

# Approach-1: Using Binary Search with rotational index
class Solution:
    def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
        # Assume 2-D matrix as a single array of size (row*col) i.e. location of target variable => row_mat = idx // col and col_mat = idx % col 
        num_rows = len(matrix)

        if (num_rows == 0):
            return False
        
        num_cols = len(matrix[0])

        l = 0
        h = (num_rows * num_cols) - 1
        
        # print(l,h)

        while (l <= h):
            mid = (l + h) // 2            
            mid_val = matrix[mid // num_cols][mid % num_cols]

            # print(mid, mid_val)

            if (mid_val == target):
                return True
            
            elif (mid_val > target):
                h = mid - 1

            else:
                # mid_val < target
                l = mid + 1

        return False

# Approach-2: Using Binary Search with two pointer concept
class Solution:
    def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
        num_rows = len(matrix)

        if (num_rows == 0):
            return False
        
        num_cols = len(matrix[0])

        # Assume we are starting from last row and first column
        
        r = num_rows - 1   # Keeps track of rows
        c = 0   # Keeps track of cols


        while(r >= 0 and c < num_cols):
            if (matrix[r][c] > target):
                r -= 1
            elif (matrix[r][c] < target):
                c += 1
            else:
                return True
        
        return False

if __name__ == "__main__":

    solution = Solution()

    # Example-1
    matrix = [[1,3,5,7],[10,11,16,20],[23,30,34,60]]
    target = 3
    # Output: True

    result = solution.searchMatrix(matrix, target)
    print(result)

    # Example-2
    matrix = [[1,3,5,7],[10,11,16,20],[23,30,34,60]]
    target = 13
    # Output: False

    result = solution.searchMatrix(matrix, target)
    print(result)
```
